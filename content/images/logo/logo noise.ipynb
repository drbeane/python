```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
n = 200
ps = 240
k = 0
a = 0.9

np.random.seed(5)
o = np.random.uniform(low=-1, high=1, size=2*n).reshape(-1,2)
plt.figure(figsize=[4,4])
plt.scatter([-1], [1])
plt.scatter(o[:,0], o[:,1], s=ps, edgecolor=(k,k,k), alpha=a)
plt.axis('off')
plt.show()
```

```python
np.random.seed(6)
b = np.random.uniform(low=-1, high=1, size=2*n).reshape(-1,2)
plt.figure(figsize=[4,4])
plt.scatter(b[:,0], b[:,1], s=ps, edgecolor=(k,k,k), alpha=a)
plt.scatter([1], [-1])
plt.axis('off')
plt.show()
```
